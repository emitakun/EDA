```python
df['installationDate'] = pd.to_datetime(df['installationDate'])
df['installation_year'] = df['installationDate'].dt.year
#df['installation_month'] = df['installationDate'].dt.month
df['installation_month'] = df['installationDate'].dt.strftime('%B')
df['installation_date'] = df['installationDate'].dt.day
df['date'] = pd.to_datetime(df['date'])
df['month'] = df['date'].dt.strftime('%B')
df['year'] = df['date'].dt.year
df['day'] = df['date'].dt.day
```

```python
df.describe()
```

```python
df.head()
```

```python
#univariate analysis
plt.hist(df['kWh'])
plt.xlabel('kWh')
```

```python
plt.hist(df['name'])
plt.xlabel('Name')
plt.xticks(rotation = 90)
```

```python
mean_generation = df['kWh'].mean()
print(mean_generation)
```

```python

#Total production by plant

Aggregate_by_plant_name = df.groupby('name')['kWh'].sum().reset_index()
print(Aggregate_by_plant_name)
```

```python
#We can see from the chart that the Bearspaw water treatment plant has the highest production. We may want to find out why those 
#with small amount of production are performing poorly. If we assume plant capacity to be thesame, then we may attribute it to lack of proper
#plant maintenance else, we will need the plant capacity for each site.
```

```python
plt.figure(figsize=(12,10))
x = Aggregate_by_plant_name['name']
y = Aggregate_by_plant_name['kWh']
plt.bar(x,y, color = 'purple')
plt.xlabel('plant_name')
plt.ylabel('Energy_produced')
plt.title('Energy_produced_by_plant')
plt.xticks(rotation = 90)
ax = plt.gca()
ax.ticklabel_format(style = 'scientific', axis = 'y', scilimits = (3,3))
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
production_by_month = df.groupby('month')['kWh'].sum().reset_index()
print(production_by_month)
```

```python
production_by_month_sorted = production_by_month.sort_values(by = 'kWh', ascending = False)
print(production_by_month_sorted)
```

```python
#July has the highest production because its the most sunny month in canada, we see that months that production is below average we will 
#require alternative sources
```

```python
x = production_by_month_sorted['month']
y = production_by_month_sorted['kWh']
plt.bar(x,y, color = 'red')
plt.ylabel('production')
plt.xlabel('month')
plt.xticks(rotation = 45)
plt.show()
```

```python
#Daily trend
daily_trend = df.groupby(['day'])['kWh'].sum().reset_index()
print(daily_trend)
```

```python
plt.figure (figsize = (12,3))
x = daily_trend['day']
y = daily_trend['kWh']
plt.title('Daily_production_trend')
plt.grid(True)
plt.plot(x,y, color = 'purple')
plt.xlabel('Day')
plt.ylabel('kWh')
plt.tight_layout()
plt.show()
```

```python
plant_age = df['year'] - df['installation_year']
print(plant_age)
```

```python
df.head()
```

```python
df['plant_age'] = df['year'] - df['installation_year']
```

```python
df.head()
```

```python
df['plant_age'].unique()
```

```python
production_by_age = df.groupby('plant_age')['kWh'].sum().reset_index()
print(production_by_age)
```

```python
plt.figure(figsize = (10,3))
plt.title('Energy_produced_by_plant_age')
x = production_by_age['plant_age']
y = production_by_age['kWh']
plt.plot(x,y, color = 'brown')
plt.xlabel('Plant_age(years)')
plt.ylabel('Energy_Generated(KWh)')
plt.tight_layout()
plt.show()
```

```python
average_production_by_site = df.groupby('name')['kWh'].mean().reset_index()
print(average_production_by_site)
```

```python
x = average_production_by_site['name']
y = average_production_by_site['kWh']
plt.figure(figsize = (12,3))
plt.barh(x,y)
plt.title('Average Production by Plant')
plt.xlabel('Energy Production by plant')
plt.ylabel('Plant_name')
plt.tight_layout()
plt.show()
```

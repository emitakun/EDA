```python
plt.figure(figsize=(12, 6))
x = df['name']
y = df['kWh']
plt.scatter(x, y, color = 'purple')
plt.xlabel('name')
plt.xticks(rotation=30, ha='right')
plt.ylabel('KWh')
plt.title('Energy Production by Plant name')
plt.tight_layout()
plt.show()
```

```python
df['date'].info()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['month_name'] = df['date'].dt.strftime('%B')
df.head()
```

```python
print(df['hour'].unique)
```

```python
print(df['year'].unique)
```

```python
#df['year'].head(50)
```

```python
df['installationDate'] = pd.to_datetime(df['installationDate'])
df['installation_year'] = df['installationDate'].dt.year
df['installation_month'] = df['installationDate'].dt.month
df['installation_day'] = df['installationDate'].dt.day
df.head()
```

```python
df['age_of_plant'] = df['year'] - df['installation_year']
df.head()
df['age_of_plant'].unique()
df['installationMonth'] = df['installationDate'].dt.strftime('%B')
df.head()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(df['age_of_plant'], df['kWh'])
plt.xlabel('age_of_plant')
plt.ylabel('KWh')
plt.title('Energy production vs plant age')
plt.tight_layout()
plt.show()
```

```python
aggregated_by_plant_name = df.groupby(['name', 'installation_year'])['kWh'].sum().reset_index()
#print(aggregated_by_plant_name)
sorted_value = aggregated_by_plant_name.sort_values(by='kWh', ascending=False)
print(sorted_value)
#sns.barplot(x='name', y='kWh', data=sorted_value)
#plt.xticks(rotation=90)
#plt.show()

```
